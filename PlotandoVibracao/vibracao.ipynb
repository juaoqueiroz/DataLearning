```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vibration_datawtime = pd.read_csv('DadosVibracao.csv')
valor_inicial = 0
delta_t = 3.906250e-5
vibration_datawtime['Tempo'] = valor_inicial + (vibration_datawtime.index * delta_t)
vibration_datawtime.to_csv('vibration_datawtime', index = False)
```

```python
sns.set(rc={'figure.figsize':(20.0,8.0)})
graf_tensao = sns.lineplot(data = vibration_datawtime, x = "Tempo", y="Untitled", color = "red")
plt.show(graf_tensao)
```
